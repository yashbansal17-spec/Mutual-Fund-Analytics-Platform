# Performance Analytics - Day 4

This notebook documents daily returns, trailing CAGR, Sharpe, Sortino, Alpha, Beta, maximum drawdown, composite fund scorecard, and benchmark tracking error.


## Formula Reference
- Daily return = `nav_t / nav_t-1 - 1`
- CAGR = `(NAV_end / NAV_start) ** (1 / n_years) - 1`
- Sharpe = `(Rp - Rf) / Std(Rp) * sqrt(252)`, with `Rf = 6.5%`
- Sortino uses downside standard deviation from negative return days only
- Alpha/Beta use `scipy.stats.linregress` against NIFTY100 daily returns; alpha = intercept * 252
- Max drawdown = `min(NAV / running_max - 1)`
- Tracking error = `std(fund_return - benchmark_return) * sqrt(252)`
