# Advanced Analytics

This notebook documents VaR/CVaR, rolling Sharpe, investor cohorts, SIP continuity, fund recommendation, and sector HHI concentration outputs.

## Method Summary

- Historical VaR 95% is the 5th percentile of daily returns.
- CVaR is the mean of daily returns below the VaR threshold.
- Rolling Sharpe uses 90 trading-day rolling mean divided by rolling standard deviation, annualised by sqrt(252).
- Cohorts are grouped by each investor's first transaction year.
- SIP continuity flags investors with 6+ SIPs and average gap above 35 days as at-risk.
- Sector HHI is the sum of squared sector weights per equity fund.

```python
import pandas as pd
from pathlib import Path

processed = Path('../data/processed')
reports = Path('../reports')
var_cvar = pd.read_csv(processed / 'var_cvar_report.csv')
cohorts = pd.read_csv(processed / 'cohort_analysis.csv')
sip = pd.read_csv(processed / 'sip_continuity.csv')
hhi = pd.read_csv(processed / 'sector_hhi.csv')
var_cvar.head()
```

## Five Advanced Insights

1. ABSL Small Cap Fund - Regular - Growth has the most negative 95% historical VaR at -2.39%, indicating the largest one-day downside threshold.
2. The 2024 investor cohort has the highest total invested amount at Rs 349.11 Cr.
3. 97.8% of investors with at least 6 SIP transactions are flagged as at-risk using the average-gap > 35 days rule.
4. Axis Bluechip Fund - Regular - Growth has the highest sector HHI at 0.297, meaning its equity portfolio is the most sector-concentrated.
