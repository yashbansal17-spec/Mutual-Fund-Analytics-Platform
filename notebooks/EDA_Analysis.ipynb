# 10 Key EDA Findings



## Finding 1

Chart 01 shows that most NAV series participated in the broad 2023 bull run, while the highlighted 2024 period created visible pauses or corrections across equity-oriented schemes.


## Finding 2

Chart 03 shows SBI Mutual Fund as the AUM leader at Rs 12.5 lakh crore in 2025-12-31, supporting the dominance callout.


## Finding 3

Chart 04 shows SIP inflows peaking at Rs 31,002 crore in Dec 2025, the all-time high in the dataset.


## Finding 4

Chart 05 shows Liquid as the strongest aggregate category by net inflow, with total inflow of Rs 451,275 crore.


## Finding 5

Chart 06 shows 26-35 as the largest investor age group in transaction records.


## Finding 6

Chart 09 shows Madhya Pradesh leading SIP contribution among states, with Rs 20,682,243 invested through SIP transactions.


## Finding 7

Chart 10 shows T30 cities contributing 66.3% of total investor transactions, indicating a meaningful metro/large-city skew.


## Finding 8

Chart 11 shows total folios rising from 13.26 crore to 26.12 crore, a 97.0% increase.


## Finding 9

Chart 12 shows the highest pairwise return correlation among selected top-AUM funds is only 0.05, suggesting limited co-movement and potential diversification value.


## Finding 10

Chart 13 shows Banking as the largest aggregate equity portfolio sector exposure across holdings.
